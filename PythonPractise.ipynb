```python
import re
name="Python is 1"
digitcount=re.sub("[^0-9]","",name)
lettercount=re.sub("[^a-zA-Z]","",name)
spacecount=re.findall("[ \s]",name)
print(len(digitcount))
print(len(lettercount))
print(len(spacecount))
```

```python
def func(word):
  count=0
  sp_char="!@#$%^&*()<>{}"
  for char in word:
    if char in sp_char:
      count+=1
  return count
print(func("Hello!@#"))
```

```python
num=int(input())
for i in range(num+1):
  for s in range(num-i):
    print(" ",end="")
  for j in range(2*i-1):
    print("*",end="")
  print()
for i in range(num-1,0,-1):
  for s in range(num-i):
    print(" ",end="")
  for j in range(2*i-1):
    print("*",end="")
  print()
```

```python
num=int(input())
for i in range(2,num):
  is_prime=True
  for j in range(2,i):
    if i%j==0:
      is_prime=False
      break
  if is_prime:
    print(i)

```

```python
count=0
number=20
n=2
while count<number:
  is_prime=True
  for i in range(2,n):
    if n%i==0:
      is_prime=False
      break
  if is_prime:
    print(n)
    count+=1
  n+=1
```

```python
terms=20
count=0
n=2
while count<terms:
  is_prime=True
  for i in range(2,n):
    if n%i==0:
      is_prime=False
      break
  if is_prime:
    print(n)
    count+=1
  n+=1
```
