```python
import re
def func(s):
  new=re.sub(r'[^A-Za-z0-9]','',s).lower()
  if new==new[::-1]:
    return True
  else:
    return False

s="tab a cat"
print(func(s))
```

```python
def func(nums,target):
  lower=0
  flag=0
  upper=len(nums)-1
  while lower<=upper:
    mid=int((lower+upper)/2)
    if nums[mid]==target:
      flag+=1
      break
    elif nums[mid]<target:
      lower=mid+1
    else:
      upper=mid-1
  if flag==1:
      return mid
  else:
    return -1

nums=[-1,0,2,4,6,8]
print(func(nums,3))
```
