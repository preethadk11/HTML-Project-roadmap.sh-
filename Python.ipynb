Hollow Square


```python
for i in range(1,6):
  for j in range(1,6):
    if i==1 or i==5 or j==1 or j==5:
      print("*",end="")
    else:
      print(" ",end="")
  print()
```

Hollow Rectangle with Row Number

```python
rows=4
cols=6
for i in range(1,4+1):
  for j in range(1,6+1):
    if i==1 or i==rows:
      print(i,end="")
    elif j==1 or j==cols:
      print(i,end="")
    else:
      print(" ",end="")
  print()
```

Hollow Rectangle with Column Number

```python
rows=4
cols=6
for i in range(1,4+1):
  for j in range(1,6+1):
    if i==1 or i==rows:
      print(j,end="")
    elif j==1 or j==cols:
      print(j,end="")
    else:
      print(" ",end="")
  print()
```

Double Border Rectangle

```python
rows=7
cols=10
for i in range(1,7+1):
  for j in range(1,10+1):
    if i<=2 or i>rows-2:
      print("*",end="")
    elif j==1 or j==2 or j==cols or j==cols-1:
      print("*",end="")
    else:
      print(" ",end="")
  print()
```

Hollow rectangle with diagonal

```python
n=6
for i in range(n+1):
  for j in range(n+1):
    border_condition=i==0 or i==n or j==0 or j==n
    diagonal_condition=i+j==n-1
    if border_condition or diagonal_condition:
      print("*",end="")

    else:
      print(" ",end="")
  print()
```

Pyramid

```python
n=9
for i in range(n):
  for j in range(n-i):
    print(" ",end="")
  print("*"*(2*i+1))
```
